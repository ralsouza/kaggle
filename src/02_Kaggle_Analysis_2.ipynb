# Analysis 2

```
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style="white")
%matplotlib inline
```

```
# Dataset
clean_data_path = "drive/My Drive/datasets/autos.csv"
df = pd.read_csv(clean_data_path,encoding="latin-1")
```

```
df.head()
```

### Número de veículos pertencentes a cada marca

```
# Crie um Plot que mostre o número de veículos pertencentes a cada marca
plt.figure(figsize=(12,8), dpi=80)
sns.set_style('whitegrid')
g = sns.countplot(y='brand',data=df)
plt.title('Count Vehicle by Brand',fontsize=18)
plt.ylabel('Brand',fontsize=13)
plt.xlabel('Count of vehicles',fontsize=13)
```

### Preço médio dos veículos com base no tipo de veículo, bem como no tipo de caixa de câmbio

```
fig, ax = plt.subplots(figsize=(8,5))
sns.barplot(x='vehicleType',y='price',hue='gearbox',data=df)
ax.set_title('Average of Vehicle Price by Brand and Gearbox',fontdict={'size':18})
ax.yaxis.set_label_text('Price',fontdict={'size':13})
ax.xaxis.set_label_text('Vehicle Type',fontdict={'size':13})
plt.show()
```
